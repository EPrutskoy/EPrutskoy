### Анализ топ-20 слов для токсичных и нетоксичных комментариев

Давайте проанализируем распределение слов в каждом классе. \
Посмотрим, какие слова встречаются чаще всего в классе 1 и классе 0.

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Функция для получения топ-слов
def get_top_words(tokens, n=20):
    all_words = [word for tokens in tokens for word in tokens]
    word_counts = Counter(all_words)
    return word_counts.most_common(n)

def preprocess_text(text):
    tokens = text.lower().split()
    tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
    return tokens
```

```python
df['tokens'] = df['text'].apply(preprocess_text)

# Разделение данных на токсичные и нетоксичные комментарии
toxic_comments = df[df['toxic'] == 1]
non_toxic_comments = df[df['toxic'] == 0]
```

```python
# Получение топ-слов для каждой группы
comments_groups = [("токсичных", toxic_comments), ("нетоксичных", non_toxic_comments)]

for label, comments in comments_groups:
    top_words = get_top_words(comments['tokens'])
    print(f"Топ-20 слов для {label} комментариев:", top_words)
```

```python
# Визуализация облака слов для токсичных комментариев
wordcloud_toxic = WordCloud(width=800, height=400, background_color='white', collocations=False).generate(' '.join([word for tokens in toxic_comments['tokens'] for word in tokens]))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_toxic, interpolation='bilinear')
plt.axis('off')
plt.title('Облако слов для токсичных комментариев')
plt.show()

# Визуализация облака слов для нетоксичных комментариев
wordcloud_non_toxic = WordCloud(width=800, height=400, background_color='white', collocations=False).generate(' '.join([word for tokens in non_toxic_comments['tokens'] for word in tokens]))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_non_toxic, interpolation='bilinear')
plt.axis('off')
plt.title('Облако слов для нетоксичных комментариев')
plt.show()
```

## Обучение BERT

### Справка о BERT

**Что такое BERT?**
BERT (Bidirectional Encoder Representations from Transformers) — это модель машинного обучения, разработанная Google для обработки естественного языка (NLP).\
В отличие от традиционных моделей, которые читают текст слева направо или справа налево, BERT обучается на полных предложениях, обрабатывая текст в обоих направлениях одновременно. \
Это позволяет модели лучше понимать контекст и значения слов в предложениях.

**Почему BERT подходит для классификации текстов?**

- **Контекстуальное понимание**: Обрабатывает текст в обоих направлениях, что позволяет понимать контекст каждого слова.
- **Предобучение на большом корпусе**: Обучен на огромных объемах текстовых данных, что позволяет модели эффективно работать с различными задачами NLP.
- **Тонкая настройка (fine-tuning)**: Модель можно адаптировать к конкретной задаче с использованием небольшого набора данных.

**Преимущества BERT:**
- **Высокая точность**: Отличные результаты на различных задачах NLP, включая классификацию текстов.
- **Гибкость**: Подходит для множества задач NLP, таких как вопросно-ответные системы, распознавание именованных сущностей, и т.д.
- **Простота использования**: Доступна в библиотеках, таких как `transformers`, что облегчает интеграцию и использование в проектах.

BERT является мощным инструментом для задач классификации текстов благодаря своему контекстуальному пониманию и способности к тонкой настройке под конкретные задачи.

### Создание кастомного датасета

**Кастомный датасет** (или пользовательский датасет) — это специализированный класс в PyTorch, \
который позволяет вам организовать данные для машинного обучения в удобной и оптимизированной форме. \
Он нужен для того, чтобы обеспечить гибкость и контроль над процессом загрузки и подготовки данных.

В нашем проекте мы создаем модель для классификации комментариев на токсичные и нетоксичные.\
Для этого нам необходимо подготовить данные таким образом, чтобы они соответствовали формату, требуемому моделью BERT.

```python
class CustomDataset(Dataset):
    def __init__(self, texts, targets, tokenizer, max_len=100):
        self.texts = texts
        self.targets = targets
        self.tokenizer = tokenizer
        self.max_len = max_len
    # Этот класс наследуется от Dataset и определяет метод __init__, который инициализирует объект класса
    def __len__(self):
        return len(self.texts)
    # Метод возвращает длину датасета, то есть количество текстов.
    def __getitem__(self, idx):
        text = str(self.texts[idx])
        target = self.targets[idx]

        encoding = self.tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=self.max_len,
            return_token_type_ids=False,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt',
        )

        return {
            'text': text,
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'targets': torch.tensor(target, dtype=torch.long)
        }
    # Получает текст и соответствующую метку по индексу idx.
    #Токенизирует текст с помощью метода encode_plus токенизатора BERT.
    #ВозвращBает словарь с текстом, ID токенов, маской внимания и меткой.
```

### Инициализация модели

Класс `BertClassifier` предназначен для классификации текста с использованием модели BERT. \
Он включает в себя методы для подготовки данных, обучения, оценки и тестирования модели, а также сохранения и загрузки модели.

Код для меня достаточно громоздкий, и легко потерять что за что отвечает, поэтому я сделал себе пометки по строкам кода. \
Также далее разберем и обоснуем наличие отдельных элементов в коде, так называемых "Helpers". \
Одним из них уже является `DataLoader`, посмотрим, какие вспомогательные элементы содержит `BertClassifier`

```python
class BertClassifier:
    def __init__(self, model_path, tokenizer_path, n_classes=2, epochs=1, model_save_path='bert.pt'):
        self.model = BertForSequenceClassification.from_pretrained(model_path, num_labels=n_classes) # Загружает предобученную модель
        self.tokenizer = BertTokenizer.from_pretrained(tokenizer_path) # и токенизатор из указанных путей.
        self.device = torch.device('cuda') # Устанавливает устройство для вычислений (cuda для использования GPU).
        self.model_save_path = model_save_path # Путь для сохранения обученной модели.
        self.max_len = 100 # Задает максимальную длину входной последовательности.
        self.epochs = epochs # Количество эпох для обучения.
        self.model.to(self.device) # Перемещает модель на GPU.

    def preparation(self, X_train, y_train, X_valid, y_valid):
        self.train_set = CustomDataset(X_train, y_train, self.tokenizer, self.max_len) # Создает тренировочный набор данных
        self.valid_set = CustomDataset(X_valid, y_valid, self.tokenizer, self.max_len) # Создает валидационный набор данных
        
        self.train_loader = DataLoader(self.train_set, batch_size=2, shuffle=True) # Создает загрузчик данных для тренировки
        self.valid_loader = DataLoader(self.valid_set, batch_size=2, shuffle=True) # Создает загрузчик данных для валидации

        self.optimizer = AdamW(self.model.parameters(), lr=2e-5, correct_bias=False) # Инициализирует оптимизатор AdamW
        self.scheduler = get_linear_schedule_with_warmup(
            self.optimizer,
            num_warmup_steps=0,
            num_training_steps=len(self.train_loader) * self.epochs # Инициализирует линейный планировщик обучения
        )
        self.loss_fn = torch.nn.CrossEntropyLoss().to(self.device) # Определяет функцию потерь

    def fit(self):
        self.model.train() # Переходит в режим обучения
        losses = []
        all_preds = []
        all_targets = []

        for data in self.train_loader: # Цикл обучения
            input_ids = data["input_ids"].to(self.device) # Перемещает input_ids на GPU
            attention_mask = data["attention_mask"].to(self.device) # Перемещает attention_mask на GPU
            targets = data["targets"].to(self.device) # Перемещает targets на GPU

            outputs = self.model( # Получает выход модели
                input_ids=input_ids,
                attention_mask=attention_mask
            )

            preds = torch.argmax(outputs.logits, dim=1) # Вычисляет предсказания
            loss = self.loss_fn(outputs.logits, targets) # Вычисляет функцию потерь

            all_preds.extend(preds.cpu().numpy()) # Собирает все предсказания
            all_targets.extend(targets.cpu().numpy()) # Собирает все цели

            losses.append(loss.item()) # Собирает все значения функции потерь
            
            loss.backward() # Обратное распространение ошибки
            torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=1.0) # Ограничивает норму градиента
            self.optimizer.step() # Шаг оптимизации
            self.scheduler.step() # Шаг планировщика
            self.optimizer.zero_grad() # Обнуляет градиенты

        train_f1 = f1_score(all_targets, all_preds) # Вычисляет F1-меру на тренировочном наборе
        train_loss = np.mean(losses) # Вычисляет среднюю потерю на тренировочном наборе
        return train_f1, train_loss

    def eval(self):
        self.model.eval() # Переходит в режим оценки
        losses = []
        all_preds = []
        all_targets = []

        with torch.no_grad(): # Без градиентов
            for data in self.valid_loader: # Цикл оценки
                input_ids = data["input_ids"].to(self.device) # Перемещает input_ids на GPU
                attention_mask = data["attention_mask"].to(self.device) # Перемещает attention_mask на GPU
                targets = data["targets"].to(self.device) # Перемещает targets на GPU

                outputs = self.model( # Получает выход модели
                    input_ids=input_ids,
                    attention_mask=attention_mask
                )

                preds = torch.argmax(outputs.logits, dim=1) # Вычисляет предсказания
                loss = self.loss_fn(outputs.logits, targets) # Вычисляет функцию потерь

                all_preds.extend(preds.cpu().numpy()) # Собирает все предсказания
                all_targets.extend(targets.cpu().numpy()) # Собирает все цели

                losses.append(loss.item()) # Собирает все значения функции потерь
        
        val_f1 = f1_score(all_targets, all_preds) # Вычисляет F1-меру на валидационном наборе
        val_loss = np.mean(losses) # Вычисляет среднюю потерю на валидационном наборе
        return val_f1, val_loss

    def train(self):
        best_f1 = 0
        for epoch in range(self.epochs): # Цикл обучения по эпохам
            print(f'Epoch {epoch + 1}/{self.epochs}')
            train_f1, train_loss = self.fit() # Обучение модели на тренировочном наборе
            print(f'Train loss {train_loss} F1-score {train_f1}')

            val_f1, val_loss = self.eval() # Оценка модели на валидационном наборе
            print(f'Val loss {val_loss} F1-score {val_f1}')
            print('-' * 10)

            if val_f1 > best_f1: # Сохранение лучшей модели
                torch.save(self.model, self.model_save_path)
                best_f1 = val_f1

        self.model = torch.load(self.model_save_path) # Загрузка лучшей модели
        
    def load_model(self, model_save_path):
        self.model.load_state_dict(torch.load(model_save_path)) # Загрузка состояния модели
        self.model.eval() # Перевод модели в режим оценки

    def test(self, X_test, y_test):
        self.model.eval() # Переходит в режим оценки
        test_set = CustomDataset(X_test, y_test, self.tokenizer, self.max_len) # Создает тестовый набор данных
        test_loader = DataLoader(test_set, batch_size=2, shuffle=False) # Создает загрузчик данных для теста

        all_preds = []
        all_targets = []

        with torch.no_grad(): # Без градиентов
            for data in test_loader: # Цикл тестирования
                input_ids = data["input_ids"].to(self.device) # Перемещает input_ids на GPU
                attention_mask = data["attention_mask"].to(self.device) # Перемещает attention_mask на GPU
                targets = data["targets"].to(self.device) # Перемещает targets на GPU

                outputs = self.model( # Получает выход модели
                    input_ids=input_ids,
                    attention_mask=attention_mask
                )

                preds = torch.argmax(outputs.logits, dim=1) # Вычисляет предсказания

                all_preds.extend(preds.cpu().numpy()) # Собирает все предсказания
                all_targets.extend(targets.cpu().numpy()) # Собирает все цели

        test_f1 = f1_score(all_targets, all_preds) # Вычисляет F1-меру на тестовом наборе
        return test_f1
```

### Подготовка данных

Мы создадим функцию, которая принимает тренировочные и тестовые наборы данных, \
применяет стемминг и лемматизацию, а затем преобразует тексты в TF-IDF признаки. \
Эта функция вернет преобразованные наборы данных, готовые для обучения и тестирования моделей машинного обучения.

Стемминг и лемматизация являются важными методами обработки текста и могут значительно улучшить качество моделей машинного обучения. \
Они помогают нормализовать слова, приводя их к базовой форме.

* **Стемминг** - Удаляет суффиксы слов, чтобы привести их к базовой форме `(например, "running" -> "run")`. \
Стемминг часто бывает "агрессивным" и может создавать несуществующие слова.

* **Лемматизация** - Приводит слова к их леммам, учитывая контекст `(например, "running" -> "run", "better" -> "good")`.\
Лемматизация точнее, но требует больше вычислительных ресурсов.

```python
# Функция очистки текста
def clean_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.lower()
    return text

# Функция получения эмбеддингов и токенов BERT
def get_bert_embeddings_and_tokens(texts, model, tokenizer, device, batch_size=200, max_len=50):
    embeddings = []
    tokens_list = []
    for i in range(0, len(texts), batch_size):
        batch_texts = texts[i:i + batch_size]
        batch_texts = [clean_text(text) for text in batch_texts]
        inputs = tokenizer(batch_texts, return_tensors='pt', padding=True, truncation=True, max_length=max_len).to(device)
        with torch.no_grad():
            outputs = model(**inputs)
        batch_embeddings = outputs.last_hidden_state.mean(dim=1).cpu().numpy()
        embeddings.extend(batch_embeddings)
        tokens = [tokenizer.convert_ids_to_tokens(ids) for ids in inputs['input_ids'].cpu().numpy()]
        tokens_list.extend(tokens)
    return np.array(embeddings), tokens_list
```

### Построение и обучение моделей Логистической регрессии, SVM и Native Bayes

Обоснование моделей в рамках проекта:

* **Логистическая регрессия**: Эффективна для бинарной классификации. \
Она хорошо работает с линейно разделимыми данными и легко интерпретируема, что делает её подходящей для базового анализа влияния признаков.

* **SVM (Support Vector Machine)**: Подходит для работы с высокоразмерными данными, такими как тексты, обработанные через TF-IDF. \
SVM эффективен при наличии чёткой разделительной границы между классами и может использовать различные ядра для решения нелинейных задач.

* **Наивный Байес (Naive Bayes)**: Основан на применении теоремы Байеса с "наивным" предположением о независимости признаков. \
Эта модель быстро обучается и часто показывает хорошие результаты в задачах текстовой классификации, особенно когда объем данных ограничен.

Поскольку процесс создания и инициализации объектов GridSearch довольно похож,\
можно обернуть процесс в функцию, чтобы избежать повторения.

```python
def perform_grid_search(model, param_grid, X_train, y_train, model_name, results_df, cv=2):
    scorer = make_scorer(f1_score)
    # Запись времени начала выполнения
    start_time = time.time()

    # Создание пайплайна
    pipeline = Pipeline([
        ('model', model)
    ])

    grid_search = GridSearchCV(pipeline, param_grid, cv=cv, scoring=scorer)
    grid_search.fit(X_train, y_train)
    
    best_params = grid_search.best_params_
    best_score = grid_search.best_score_

    end_time = time.time()
    duration = end_time - start_time

    print(f"Лучшие параметры для {model_name}: {best_params}")
    print(f"Лучший F1-score на кросс-валидации для {model_name}: {best_score}")
    
    # Сохранение результатов в DataFrame
    result = pd.DataFrame({
        'Model Name': [model_name],
        'Best Parameters': [best_params],
        'Best F1-Score': [best_score],
        'Time': [duration]
    })
    results_df = pd.concat([results_df, result], ignore_index=True)
    
    return results_df
```

```python
# Создание пустой таблицы для результатов
results_df = pd.DataFrame(columns=['Model Name', 'Best Parameters', 'Best F1-Score', 'Time'])
```

```python
param_grid_lr = {'model__C': [0.1, 1, 10], 'model__solver': ['liblinear', 'lbfgs']}
log_reg = LogisticRegression(max_iter=1500, randoma_state=42)
results_df = perform_grid_search(log_reg, param_grid_lr, X_train, y_train, 'Logistic Regression', results_df)
```

```python
param_grid_svm = {'model__C': [0.1, 0.01], 'model__kernel': ['linear']}
svm = SVC(randoma_state=42)
results_df = perform_grid_search(svm, param_grid_svm, X_train, y_train, 'SVM', results_df)
```

Использование функции GridSearchCV позволило эффективно настроить параметры для моделей Логистической регрессии, SVM и Naive Bayes,\
что значительно упрощает процесс оптимизации моделей. \
Логистическая регрессия с параметрами {'C': 10, 'solver': 'liblinear'} показала наилучший результат с F1-счетом 0.871596, что делает её предпочтительным выбором для данной задачи классификации.\
